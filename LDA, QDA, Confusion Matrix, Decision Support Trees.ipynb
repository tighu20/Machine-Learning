```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
DataCancer=load_breast_cancer()
#print(DataCancer.DESCR)
print(type(DataCancer.feature_names))
#print(DataCancer.keys())
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
X_features=DataCancer.data
Y_targetClass=DataCancer.target
X_train, X_test, Y_train, Y_test = train_test_split(X_features,Y_targetClass,random_state=0)

for x in [0.1, 1,5, 10,50, 100,1000]:
    LogRegModel= LogisticRegression(C=x).fit(X_train, Y_train)
    #print("the accuracy of   model  with c=",x,"is",LogRegModel.score(X_test,Y_test))
    #print(LogRegModel.coef_.shape)
    LogRegModelcoef=LogRegModel.coef_.reshape(30)
    dataFrame1=pd.DataFrame({'feature_names':DataCancer.feature_names, 'Estimated Coefficints':LogRegModelcoef }) 
    print("Feature names and Thier coefficients when c = ",x) 
    display(dataFrame1)
   
    
    
    
```

As we increase C the regularisation strength decreases. Low C means there will be significant numbers of coefficients which will be zero as can be seen from the tables above. 

The flexiblity of models increases as we increase C because we are effectively increases the number of features from which model will be trained. High C means more features which is equivalent to a more flexible model.
